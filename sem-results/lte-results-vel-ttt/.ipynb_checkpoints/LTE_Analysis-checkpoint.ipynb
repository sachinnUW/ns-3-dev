# Imports

```python
import sys
sys.path.append("../../sem")
import sem
import pprint
import matplotlib.pyplot as plt
import seaborn as sns
import numpy
sns.set_style("white")
from sklearn.linear_model import LinearRegression
```

```python
# Open SEM campaign with the required path & directory.
campaign_dir = "../lte-results-vel-ttt"
campaign = sem.CampaignManager.load(campaign_dir, check_repo=False)
```

# Or Define New SEM Campaign

```python
# Create new SEM campaign with the required path & directory.
ns_path = '../../'
script = 'lena-handover-many'
campaign_dir = "../lte-results-vel-ttt"
campaign = sem.CampaignManager.new(ns_path, script, campaign_dir, overwrite=True, check_repo=False)
```

# Run the Simulation

```python
# Run the missing simulations with the parameters.
params = {'hysteresis': 3, 'timeToTrigger': list([0.0, 40.0, 64.0, 80.0, 100.0, 128.0, 160.0, 256.0, 320.0, 480.0, 512.0, 640.0, 1024.0, 1280.0, 2560.0, 5120.0]), 'speed': list([10.0, 20.0, 30.0, 40.0, 50.0, 60.0, 70.0, 80.0, 90.0, 100.0, 110.0, 120.0, 130.0, 140.0, 150.0, 160.0]), 'simTime': [5], 'useUdp': [1], 'RngRun': 1}
campaign.run_missing_simulations(params, stop_on_errors=False)
```

# Parse the Results

```python
# Function to get average throughput, number of HOs and RLFs
def get_average_throughput(result):    
    if result['output']['stderr'] != "":
        throughput = -1
        num_HOs = -1
        num_RLFs = -1
    else:
        split_results = result['output']['stdout'].split("\n")
        throughput = float(split_results[-1].split(" ")[1])
        num_HOs = 0
        num_RLFs = 0
        for line in split_results:
            event = line.split(":")[0]
            if event == "UE_HO_START":
                num_HOs += 1
            elif event == "UE_RLF_DETECTED":
                num_RLFs += 1	
    return [num_HOs, num_RLFs, throughput]
```

```python
# Use the parsing function to create a Pandas dataframe
results = campaign.get_results_as_dataframe(get_average_throughput, ['Num_HOs', 'Num_RLFs', 'Throughput'], params=params, drop_constant_columns=True)
print(results)
```

```python
fig = plt.figure(figsize=(25, 15))
ax = plt.axes(projection='3d')
ax.plot_trisurf(results['timeToTrigger'], results['speed'], results['Throughput'])
ax.set_xlabel('Time To Trigger')
ax.set_ylabel('Velocity')
ax.set_zlabel('Throughput')
ax.set_title('Throughput Plot for Vel-TTT')
```

```python
fig = plt.figure(figsize=(25, 15))
ax = plt.axes(projection='3d')
ax.plot_trisurf(results['timeToTrigger'], results['speed'], results['Num_HOs'])
ax.set_xlabel('Time To Trigger')
ax.set_ylabel('Velocity')
ax.set_zlabel('Num_HOs')
ax.set_title('Num_HOs Plot for Vel-TTT')
```

```python
fig = plt.figure(figsize=(25, 15))
ax = plt.axes(projection='3d')
ax.plot_trisurf(results['timeToTrigger'], results['speed'], results['Num_RLFs'])
ax.set_xlabel('Time To Trigger')
ax.set_ylabel('Velocity')
ax.set_zlabel('Num_RLFs')
ax.set_title('Num_RLFs Plot for Vel-TTT')
```

```python
results['speed*ttt'] = results['timeToTrigger']*results['speed']
print(results)
```

```python
results2 = results[results['speed*ttt'] <= 200000.0]
regr = LinearRegression()
regr.fit(results2['speed*ttt'].to_numpy().reshape(-1, 1), results2['Num_RLFs'])
pred = regr.predict(results2['speed*ttt'].to_numpy().reshape(-1, 1))
```

```python
fig = plt.figure(figsize=(6, 4))
ax = plt.axes()
ax.scatter(results2['speed*ttt'], results2['Num_RLFs'])
ax.plot(results2['speed*ttt'], pred, color = 'red')
ax.grid()
ax.set_xlabel('speed*timeToTrigger')
ax.set_ylabel('Number of RLFs')
ax.set_title('Number of RLFs Plot for speed*TTT')
```

```python
results2 = results[(results['timeToTrigger'] == 1024) & (results['speed'] <=120)]
fig = plt.figure(figsize=(10, 4))
ax = plt.axes()
ax.plot(results2['speed'], results2['Num_HOs'])
ax.plot(results2['speed'], results2['Num_RLFs'], color = 'red')
ax.grid()
ax.set_xlabel('speed')
ax.set_ylabel('Number of HOs and RLFs')
ax.set_title('Number of HOs and RLFs Plot for speed (a3Off = 0 dB, TTT = 1024 ms, Hyst = 3 dB)')
ax.legend(['Number of HOs', 'Number of RLFs'])
```
